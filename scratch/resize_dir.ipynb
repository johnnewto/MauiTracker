```python

#!/usr/bin/python
from PIL import Image
import os, sys
from cv2 import cv2
from pathlib import Path
home = str(Path.home())
path_in = home+"/data/Karioitahi_09Feb2022/132MSDCF-28mm-f4/"
path_out = home+"/data/Karioitahi_09Feb2022/132MSDCF-28mm-f4-small/"

dirs = os.listdir( path_in )
```

```python
def resize():
    for item in dirs:
        if os.path.isfile(path_in+item):
            im = Image.open(path_in+item)
            f, e = os.path.splitext(path_in+item)
            imResize = im.resize((480,320), Image.ANTIALIAS)
            imResize.save(path_out + item, 'JPEG', quality=90)
            print(path_out + item)

resize()
```
